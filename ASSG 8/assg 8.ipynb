```python
def testmatch(img):
    img0 = cv2.imread(img)
    img0 = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
    plt.imshow(img0,cmap="gray")
    keypoints_0, descriptors_0 = sift.detectAndCompute(img0,None)
    #print(len(keypoints_0))

    matchesA = bf2.knnMatch(descriptorsA,descriptors_0, k=2)
    # Apply ratio test
    goodA = []
    for m,n in matchesA:
        if m.distance < 0.75*n.distance:
            goodA.append([m])
    print(len(goodA))
    print(len(matchesA))
    
    matchesB = bf2.knnMatch(descriptorsB,descriptors_0, k=2)
    # Apply ratio test
    goodB = []
    for m,n in matchesB:
        if m.distance < 0.75*n.distance:
            goodB.append([m])
    print(len(goodB))
    print(len(matchesB))

    matchesC = bf2.knnMatch(descriptorsC,descriptors_0, k=2)
    # Apply ratio test
    goodC = []
    for m,n in matchesC:
        if m.distance < 0.75*n.distance:
            goodC.append([m])
    print(len(goodC))
    print(len(matchesC))
    
    matchesD = bf2.knnMatch(descriptorsD,descriptors_0, k=2)
    # Apply ratio test
    goodD = []
    for m,n in matchesD:
        if m.distance < 0.75*n.distance:
            goodD.append([m])
    print(len(goodD))
    print(len(matchesD))
    
    if(len(goodA)>len(goodB) and len(goodA)>len(goodC) and len(goodA)>len(goodD) and len(goodA)>120):
        print("\n\nPersonA")
    elif(len(goodB)>len(goodC) and len(goodB)>len(goodD) and len(goodB)>125):
        print("\n\nPersonB")
    elif(len(goodC)>len(goodD) and len(goodC)>125):
        print("\n\nPersonC")
    elif(len(goodD)>125):
        print("\n\nPersonD")
    else:
        print("\n\nCan't identify")
    
```
