```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

PART A

```python
im = cv2.imread("thanos.jpg")
```

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.subplot(131),plt.imshow(im)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(gray,cmap = 'gray')
plt.title('Gray Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(binary,cmap = 'gray')
plt.title('Binary Image'), plt.xticks([]), plt.yticks([])
plt.show()

grayAs3 = cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
binaryAs3 = cv2.cvtColor(binary,cv2.COLOR_GRAY2BGR)
all_imgs = np.concatenate((im,grayAs3,binaryAs3), axis=1)
cv2.imshow('Orginal -Grey-binary ',all_imgs)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

PART B

```python
print("For orginal image")

max_vals = [np.amax(im[:,:,0]), np.amax(im[:,:,1]), np.amax(im[:,:,2])]
min_vals=[np.amin(im[:,:,0]), np.amin(im[:,:,1]), np.amin(im[:,:,2])]
print("MAX values for blue, green, red = ",max_vals)
print("MIN values for blue, green, red = ",min_vals)


```

```python
print("For binary image")

min_ = np.amin(binary)
max_ = np.amax(binary)
print("MAX value = ",max_)
print("MIN value = ",min_)
```

```python
print("For grey image")

min_ = np.amin(gray)
max_ = np.amax(gray)
print("MAX value = ",max_)
print("MIN value = ",min_)
```
